```python
# The law= |p1-p0|*|p-p0| / |p1-p0| 
# Equal cross_product/ line_length=Parallelogram_area/Base_length=distance
p = np.array([2.0, 1.0])

P0 = np.array([[0.0, 0.0], [1.0, 2.0], [0.0, 1.0]])

P1 = np.array([[4.0, 0.0], [4.0, 2.0], [3.0, 5.0]])
vec_line=P1-P0
vec_p=p-P0
cross_prod=np.abs(vec_line[:,0]*vec_p[:,1]-vec_line[:,1]*vec_p[:,0]) #🔴
line_len=np.linalg.norm(vec_line,axis=1) #🔴
distances=cross_prod/line_len
print(f"Distance from the point for every line---> {distances}")
```

```python
P0 = np.array([[0.0, 0.0], [1.0, 1.0]])  
P1 = np.array([[4.0, 0.0], [4.0, 3.0]])  

P = np.array([[2.0, 1.0], [0.0, 2.0], [3.0, 3.0]])
P0_bro=P0[np.newaxis,:,:] #shape: (1,N_lines,2)
P1_bro = P1[np.newaxis, :, :]  #shape: (1, N_lines, 2)
P_bro  = P[ :, np.newaxis, :]  #shape: (M_points, 1, 2)
vec_line = P1_bro - P0_bro          
vec_p = P0_bro - P_bro
cross_prod = np.abs(vec_line[:, :, 0] * vec_p[:, :, 1] - vec_line[:, :, 1] * vec_p[:, :, 0])

line_length = np.linalg.norm(vec_line, axis=2) 
distance=cross_prod/line_length
print(distance)
```

```python
def game_of_life(grid, num_iterations=5):
    for _ in range(num_iterations):
        n = np.zeros_like(grid, dtype=int) #🔴
        for dr in [-1,0,1]:
            for dc in [-1,0,1]:
                if dr==0 and dc==0:
                    continue
                n+=np.roll(np.roll(grid,dr,axis=0),dc,axis=1)
        birth = (n == 3) & (grid == 0)
        survive = ((n == 2) | (n == 3)) & (grid == 1)
        grid = (birth | survive).astype(int)
    return grid
Z = np.random.randint(0, 2, (5, 5))
print(game_of_life(Z, 2))
```

```python
from scipy.interpolate import interp1d
X = np.array([0.0, 1.0, 2.5, 4.0, 7.0])
Y = np.array([0.0, 2.0, 1.0, 5.0, 3.0])
dx=np.diff(X)
dy=np.diff(Y)
segment_lengths = np.sqrt(dx**2 + dy**2)
distance=np.insert(np.cumsum(segment_lengths),0,0.0) #🔴🔴
total_length = distance[-1]
num_samples = 10
equal_distances = np.linspace(0, total_length, num_samples)
interp_x=interp1d(distance, X)
interp_y=interp1d(distance, Y)
new_X = interp_x(equal_distances)
new_Y = interp_y(equal_distances)
print("Original X:", X)
print("New Equidistant X:", new_X)
print("\nOriginal Y:", Y)
print("New Equidistant Y:", new_Y)
```

```python
np.random.seed(42)
X = np.array([12, 15, 14, 10, 18, 20, 22, 19, 11, 16])
N=1000
n_samples=len(X)
resampled_indices=np.random.randint(0,n_samples,size=(N,n_samples))
resampled_arrays=X[resampled_indices]
bootstrap_means=np.mean(resampled_arrays,axis=1)
ci_lower, ci_upper = np.percentile(bootstrap_means, [2.5, 97.5])
print(f"Original Mean: {np.mean(X):.4f}")
print(f"Bootstrapped 95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
# StandWithPalestine 🍉
```
